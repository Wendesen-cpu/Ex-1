```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('jawar.png')
img.shape
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(img, cmap="gray")
```

```python
def ToGray(img):
  row,col,dim = img.shape
  image= np.zeros((row,col))

  for i in range(row):
    for j in range(col):
      image[i,j]= 0.299*img[i,j,0] + 0.587*img[i,j,1] + 0.114*img[i,j,2]
  return np.uint8(image)
```

```python
img_gray = ToGray(img)
```

```python
plt.imshow(img_gray, cmap="gray")
plt.show()
```

```python
img_gray = cv2.GaussianBlur(img_gray,(5,5),0)
```

```python
plt.imshow(img_gray, cmap="gray")
plt.show()
```

```python
img_fft = cv2.dft(np.float32(img_gray), flags=cv2.DFT_COMPLEX_OUTPUT)
img_fft_shifted = np.fft.fftshift(img_fft)
```

```python
img_fft_shifted.shape
```

```python
magnitude_fft = 20*np.log(cv2.magnitude(img_fft_shifted[:,:,0],img_fft_shifted[:,:,1]))
```

```python
plt.imshow(magnitude_fft, cmap="gray")
```

```python
row,col = img_gray.shape
crow,ccol = int(row/2),int(col/2)
```

```python
mask = np.ones((row,col,2), dtype=np.uint8)
r =100
center = [crow,ccol]
row
```

```python
x,y = np.ogrid[:row,:col]
```

```python
mask_area = (x-center[0])**2 + (y-center[1])**2 <= r*r
```

```python
img_fft_shifted.dtype
```

```python
mask[mask_area]=0
```

```python
masked_dft = img_fft_shifted * mask
magnitude_fft = 2000*np.log(cv2.magnitude(masked_dft[:,:,0],masked_dft[:,:,1]))
```

```python
plt.imshow(magnitude_fft, cmap="gray")

```

```python
reversed_shift = np.fft.ifftshift(masked_dft)
reversed_img = cv2.idft(reversed_shift)
```

```python
magnitude_reversed_img = 20*np.log(cv2.magnitude(reversed_img[:,:,0],reversed_img[:,:,1]))
```

```python
plt.imshow(magnitude_reversed_img, cmap="gray")
```
